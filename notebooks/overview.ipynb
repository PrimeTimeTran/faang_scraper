```python
import pandas as pd
import matplotlib.pyplot as plt

path = '../tmp/downloads/2024-06_google_jobs.csv'

df = pd.read_csv(path)
```

```python
filtered_df = df[df['experience'] != 0]
average_experience_requested = filtered_df['experience'].mean()
print(f'Average Experience: {average_experience_requested}')
```

```python
location_counts = df['location'].value_counts()

top_20_locations = location_counts.sort_values(ascending=False).head(20)
plt.figure(figsize=(10, 8))
top_20_locations.plot(kind='barh')
plt.title('Top 20 Locations by Number of Jobs')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.gca().invert_yaxis()


for index, value in enumerate(top_20_locations):
    plt.text(value - 5, index, f'{value}', va='center', ha='right', color='white')
    

ax = plt.subplot()
ax.spines[['right', 'top']].set_visible(False)

plt.show()    
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and 'location' is the column of interest
location_counts = df['location'].value_counts()

# Sort locations by number of jobs in ascending order and select bottom 20
bottom_20_locations = location_counts.sort_values(ascending=True).head(20)

# Plotting the bottom 20 locations
plt.figure(figsize=(10, 8))
bars = bottom_20_locations.plot(kind='barh')
plt.title('Bottom 20 Locations by Number of Jobs')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest count at the top

# Annotate each bar with the count value on the left side
for index, value in enumerate(bottom_20_locations):
    plt.text(value - .01, index, f'{value}', va='center', ha='right', color='white')
ax = plt.subplot()
ax.spines[['right', 'top']].set_visible(False)
plt.show()
```

```python
import ast


def parse_and_extract_first(x):
    try:
        parsed_list = ast.literal_eval(x)
        if isinstance(parsed_list, list) and len(parsed_list) > 0:
            return parsed_list[0]
        else:
            return None
    except (SyntaxError, ValueError):
        return None


# Apply lambda function to extract the first element of each list in 'preferred'
df = pd.read_csv(path)
df['preferred_first'] = df['preferred'].apply(parse_and_extract_first)
preferred_counts = df['preferred_first'].value_counts()

top_30_titles = preferred_counts.sort_values(ascending=False)
plt.figure(figsize=(10, 200))
top_30_titles.plot(kind='barh')
plt.title('Top 20 Locations by Number of Jobs')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.gca().invert_yaxis()

ax = plt.subplot()
ax.spines[['right', 'top']].set_visible(False)
plt.show()


```
